```python
# RUN THIS METHOD TO READ FILES
def readP300(pathname):
    myLayout = mne.channels.make_standard_montage('standard_1020')

    scale = 1

    file_iter = glob.glob(pathname)

    bci2kfile = b2k.BCI2kReader(file_iter[0])

    info = mne.create_info(ch_names=bci2kfile.parameters['ChannelNames'], sfreq=bci2kfile.parameters['SamplingRate'],
                           ch_types=['eeg'] * bci2kfile.parameters['SourceCh'])
    raw = mne.io.RawArray(bci2kfile.signals[0:bci2kfile.parameters['SourceCh'],:]*scale, info)
    raw.set_montage(myLayout)
    bci2kfile.states['StimulusCode'][0][np.where(bci2kfile.states['StimulusType'] > 0)[1]] = -1
    bci2kfile.states['StimulusCode'][0][np.where(bci2kfile.states['StimulusCode'] > 0)[1]] = 2
    bci2kfile.states['StimulusCode'][0][np.where(bci2kfile.states['StimulusCode'] == -1)[1]] = 1


    events = np.transpose(np.array([np.where(np.diff(bci2kfile.states['StimulusCode'])>0)[1]+1,
                                    np.zeros((np.where(np.diff(bci2kfile.states['StimulusCode'])>0)[1]+1).shape),
                                    bci2kfile.states['StimulusCode'][0][np.where(np.diff(bci2kfile.states['StimulusCode'])>0)[1]+1]],
                                   dtype=int))
    print(events.shape)
    for filename in file_iter[1:]:
        print(filename)
        bci2kfile = b2k.BCI2kReader(filename)
        info = mne.create_info(ch_names=bci2kfile.parameters['ChannelNames'], sfreq=bci2kfile.parameters['SamplingRate'],
                               ch_types=['eeg'] * bci2kfile.parameters['SourceCh'])
        rawTemp = mne.io.RawArray(bci2kfile.signals[0:bci2kfile.parameters['SourceCh'],:]*scale, info)
        rawTemp.set_montage(myLayout)
        bci2kfile.states['StimulusCode'][0][np.where(bci2kfile.states['StimulusType'] > 0)[1]] = -1
        bci2kfile.states['StimulusCode'][0][np.where(bci2kfile.states['StimulusCode'] > 0)[1]] = 2
        bci2kfile.states['StimulusCode'][0][np.where(bci2kfile.states['StimulusCode'] == -1)[1]] = 1

        eventsTemp = np.transpose(np.array([np.where(np.diff(bci2kfile.states['StimulusCode']) > 0)[1] + 1 + raw.n_times,
                                            np.zeros((np.where(np.diff(bci2kfile.states['StimulusCode']) > 0)[1] + 1).shape),
                                            bci2kfile.states['StimulusCode'][0][
                                                np.where(np.diff(bci2kfile.states['StimulusCode']) > 0)[1] + 1]],
                                           dtype=int))
        events = np.append(events,eventsTemp,axis=0)
        raw.append(rawTemp)



    raw.filter(1,20, fir_design='firwin')
    return raw,events
```

```python
data, events  = readP300('/content/sample_data/demoS001R01_03-Aug-22.dat')
```

```python
def getEpochs(raw, events):
    event_id = dict(target=1, nontarget=2)
    tmin = -0.2+1/500
    tmax = 1.0

    epochs = mne.Epochs(raw, events=events, event_id=event_id, tmin=tmin,
                        tmax=tmax, baseline=None, verbose=False, preload=True)

    return epochs
```

```python
epochs = getEpochs(data,events)
```
